### Check the pattern of 2 * 3 * 30 * 30 CSI

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
csi = np.load(f"../dataset/0509/make14/01_dyn_asmb1_len30_csi.npy", mmap_mode='r')
```

```python
print(csi.shape)
```

```python
tmp = csi[:201].reshape(-1, 2, 3, 30, 30)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = (20, 12)
plt.rcParams["figure.titlesize"] = 35
plt.rcParams['lines.markersize'] = 10
plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.labelsize'] = 30
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20

for x in range(5):
    fig = plt.figure()
    fig.suptitle("Amplitude")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(tmp[j + 6 * x, 0, i , ...])
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
from scipy import signal
tmp_ = tmp[:, 0, ...] * np.exp(1.j * tmp[:, 1, ...])

tmp_r = signal.savgol_filter(np.real(tmp_), 21, 3, axis=-1) # denoise for real part
tmp_i = signal.savgol_filter(np.imag(tmp_), 21, 3, axis=-1) # denoise for imag part
tmp__ = tmp_r + 1.j * tmp_i
```

```python
print(tmp__.shape)
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Amplitude-filtered")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.abs(tmp__[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Phase")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(tmp[j + 6 * x, 1, i, ...])
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Phase-filtered")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.angle(tmp__[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Real")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.real(tmp_[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Real-filtered")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.real(tmp__[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Imag")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.imag(tmp_[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
for x in range(3):
    fig = plt.figure()
    fig.suptitle("Imag-filtered")
    fig.constrained_layout = True
    sub = fig.subplots(3, 6)
    for i in range(3):
        for j in range(6):
            sub[i , j].imshow(np.imag(tmp__[j + 6 * x, i , ...]))
            sub[i , j].set_title(f"img{j + 6 * x}_ch{i}")
    plt.show()
```

```python
plt.rcParams['figure.figsize'] = (32, 18)
fig = plt.figure()
#fig.tight_layout(rect=[0, 0.03, 1, 0.95])
subfig = fig.subfigures(1, 8)

subp = subfig[0].subplots(3, 1)
for i in range(3):
    subp[i].imshow(tmp[0, 0, i, ...])
    subp[i].set_title(f"img{0}_ch{i}")
subfig[0].suptitle("Amp", y=0.93)

subfig[1].suptitle("Amp_filtered", y=0.93)
subp = subfig[1].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.abs(tmp__[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")

subfig[2].suptitle("Phase", y=0.93)
subp = subfig[2].subplots(3, 1)
for i in range(3):
    subp[i].imshow(tmp[0, 1, i, ...])
    subp[i].set_title(f"img{0}_ch{i}")
    
subfig[3].suptitle("Phase_filtered", y=0.93)
subp = subfig[3].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.angle(tmp__[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")
 
subfig[4].suptitle("Real", y=0.93)
subp = subfig[4].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.real(tmp_[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")
    
subfig[5].suptitle("Real-filtered", y=0.93)
subp = subfig[5].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.real(tmp__[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")
  
subfig[6].suptitle("Imag", y=0.93)
subp = subfig[6].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.imag(tmp_[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")
    
subfig[7].suptitle("Imag-filtered", y=0.93)
subp = subfig[7].subplots(3, 1)
for i in range(3):
    subp[i].imshow(np.imag(tmp__[0, i, ...]))
    subp[i].set_title(f"img{0}_ch{i}")

fig.suptitle("CSI Pattern")
plt.show()
```
