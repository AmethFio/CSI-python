```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
csi = np.load('../dataset/0509/make14-finished/csi_f.npy', mmap_mode='r')
c_img = np.load('../dataset/0509/make14-finished/c_img.npy', mmap_mode='r')
r_img = np.load('../dataset/0509/make14-finished/r_img.npy', mmap_mode='r')
csi_ = csi.reshape(-1, 6 * 30 * 30)
c_img_ = c_img.reshape(-1, 128 * 128)
r_img_ = r_img.reshape(-1, 128 * 226)
```

```python
CSI_tsne = TSNE(n_components=2, random_state=33).fit_transform(csi_)
CIMG_tsne = TSNE(n_components=2, random_state=33).fit_transform(c_img_)
RIMG_tsne = TSNE(n_components=2, random_state=33).fit_transform(r_img_)
ind = np.arange(csi_.shape[0])
```

```python
%matplotlib inline
plt.style.use("dark_background")
plt.figure(figsize=(12, 9))
plt.scatter(CSI_tsne[:, 0], CSI_tsne[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("CSI")
plt.show()
plt.figure(figsize=(12, 9))
plt.scatter(CIMG_tsne[:, 0], CIMG_tsne[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("C_IMG")
plt.show()
plt.figure(figsize=(12, 9))
plt.scatter(RIMG_tsne[:, 0], RIMG_tsne[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("R_IMG")
plt.show()
```

```python
CSI_pca = PCA(n_components=2, random_state=33).fit_transform(csi_)
CIMG_pca = PCA(n_components=2, random_state=33).fit_transform(c_img_)
RIMG_pca = PCA(n_components=2, random_state=33).fit_transform(r_img_)
ind = np.arange(csi_.shape[0])
```

```python
%matplotlib inline
plt.style.use("dark_background")
plt.figure(figsize=(12, 9))
plt.scatter(CSI_pca[:, 0], CSI_pca[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("CSI")
plt.show()
plt.figure(figsize=(12, 9))
plt.scatter(CIMG_pca[:, 0], CIMG_pca[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("C_IMG")
plt.show()
plt.figure(figsize=(12, 9))
plt.scatter(RIMG_pca[:, 0], RIMG_pca[:, 1],  alpha=0.6, c=ind, cmap=plt.cm.get_cmap('rainbow', csi_.shape[0]))
plt.title("R_IMG")
plt.show()
```
